## Chapter 2 — Python Operators

```python
# --- Arithmetic Operators ---
print(10 // 3)   # floor division → 3
print(10 % 3)    # modulus       → 1
print(2 ** 10)   # exponent      → 1024
print(-7 // 2)   # ⚠️ Trap: floor towards -∞ → -4, not -3
```

```python
# --- Comparison & Logical ---
print(0 == False)    # True  (value equality)
print(0 is False)    # False (identity) ← Borderline
print(not 0)         # True
print(1 < 2 < 3)     # True  — chained comparison
```

```python
# --- Bitwise ---
print(5 & 3)   # 1
print(5 | 3)   # 7
print(5 ^ 3)   # 6
print(~5)      # -6  ⚠️ Trap: ~n == -(n+1)
```

```python
# --- Walrus := (assignment expression) ---
import re
if m := re.search(r'\d+', 'item99'):
    print('Found:', m.group())   # 99
```

## Chapter 4 — Flow Control

```python
# --- if / elif / else ---
x = 0
if x > 0:   print('positive')
elif x < 0: print('negative')
else:        print('zero')

# Ternary
result = 'even' if x % 2 == 0 else 'odd'
print(result)
```

```python
# --- for / while / break / continue / else ---
for i in range(5):
    if i == 3: break
else:
    print('no break')  # ⚠️ Trap: else runs only if loop completed without break

i = 0
while i < 3:
    i += 1
print('while done, i =', i)
```

```python
# --- match-case (Python 3.10+) ---
command = 'quit'
match command:
    case 'quit': print('Exiting')
    case 'help': print('Help menu')
    case _:      print('Unknown')
```

## Chapter 6 — List Data Structure

```python
# --- Mutable, ordered, duplicates allowed ---
lst = [3, 1, 4, 1, 5, 9]
lst.append(2)
lst.insert(0, 0)
print(lst)
print(lst.index(1))  # first occurrence
```

```python
# --- List comprehension ---
squares = [x**2 for x in range(10) if x % 2 == 0]
print(squares)  # [0, 4, 16, 36, 64]
```

```python
# ⚠️ Trap: shallow copy
a = [1, [2, 3]]
b = a.copy()   # shallow
b[1].append(4)
print(a)  # [1, [2, 3, 4]] — nested list mutated!

import copy
c = copy.deepcopy(a)
c[1].append(99)
print(a)  # unchanged
```

```python
# Borderline: list * n is shallow
mat = [[0]*3]*3
mat[0][0] = 9
print(mat)  # all rows changed! use [[0]*3 for _ in range(3)]
```

## Chapter 7 — Tuple Data Structure

```python
# --- Immutable, ordered ---
t = (1, 2, 3, 2)
print(t.count(2))  # 2
print(t.index(3))  # 2
a, b, *rest = t    # unpacking
print(a, b, rest)
```

```python
# ⚠️ Trap: single-element tuple needs trailing comma
wrong = (1)   # int!
correct = (1,) # tuple
print(type(wrong), type(correct))

# Borderline: tuple with mutable element is 'immutable reference'
t2 = (1, [2, 3])
t2[1].append(4)  # allowed! list inside is mutable
print(t2)
```

## Chapter 10 — Functions

```python
# --- def, args, kwargs, defaults ---
def greet(name, msg='Hello'):
    return f'{msg}, {name}!'

print(greet('Musk'))
print(greet('Musk', msg='Hi'))
```

```python
# --- *args and **kwargs ---
def show(*args, **kwargs):
    print(args, kwargs)

show(1, 2, 3, x=10, y=20)
```

```python
# --- lambda, map, filter, reduce ---
from functools import reduce
nums = [1,2,3,4,5]
print(list(map(lambda x: x**2, nums)))
print(list(filter(lambda x: x%2==0, nums)))
print(reduce(lambda a,b: a+b, nums))   # 15
```

```python
# ⚠️ Trap: mutable default argument
def append_to(val, lst=[]):
    lst.append(val)
    return lst

print(append_to(1))  # [1]
print(append_to(2))  # [1, 2] ← shared across calls!

# Fix:
def append_fixed(val, lst=None):
    if lst is None: lst = []
    lst.append(val)
    return lst
```

```python
# --- Closures & nonlocal ---
def counter():
    count = 0
    def inc():
        nonlocal count
        count += 1
        return count
    return inc

c = counter()
print(c(), c(), c())  # 1 2 3
```

## Chapter 13 — File Handling

```python
# --- Read / Write / Append ---
import tempfile, os

# Write
with tempfile.NamedTemporaryFile(mode='w', suffix='.txt', delete=False) as f:
    fname = f.name
    f.write('Line 1\nLine 2\nLine 3\n')

# Read all
with open(fname) as f:
    print(f.read())

# Read line by line
with open(fname) as f:
    for line in f:
        print(repr(line.strip()))

os.unlink(fname)
```

```python
# ⚠️ Trap: forgetting to close — always use 'with'
# Bad:  f = open('x.txt'); data = f.read()  # may not close on exception
# Good: with open('x.txt') as f: data = f.read()

# Borderline: 'r' mode raises FileNotFoundError; 'a' creates file
try:
    open('nonexistent.txt')
except FileNotFoundError as e:
    print(e)
```

## Chapter 14 — Exception Handling

```python
# --- try / except / else / finally ---
def safe_div(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return 'Cannot divide by zero'
    except TypeError as e:
        return f'Type error: {e}'
    else:
        return result          # runs only if no exception
    finally:
        pass                   # always runs

print(safe_div(10, 2))   # 5.0
print(safe_div(10, 0))   # error message
print(safe_div(10,'x'))  # type error
```

```python
# --- Custom exceptions ---
class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        super().__init__(f'Need {amount}, have {balance}')

try:
    raise InsufficientFundsError(100, 50)
except InsufficientFundsError as e:
    print(e)
```

```python
# ⚠️ Trap: bare except catches everything including KeyboardInterrupt
try:
    pass
except:        # avoid this — catches SystemExit too
    pass

# Better:
try:
    pass
except Exception:
    pass

# Borderline: exception chaining
try:
    int('x')
except ValueError as e:
    raise RuntimeError('parse failed') from e
```

## Chapter 15 — OOP (Object-Oriented Programming)

```python
# --- Class, __init__, methods ---
class BankAccount:
    interest_rate = 0.05  # class variable

    def __init__(self, owner, balance=0):
        self.owner   = owner       # instance variable
        self._balance = balance    # 'protected' convention

    def deposit(self, amount):
        self._balance += amount

    def withdraw(self, amount):
        if amount > self._balance:
            raise ValueError('Insufficient funds')
        self._balance -= amount

    @property
    def balance(self): return self._balance

    @classmethod
    def set_rate(cls, rate): cls.interest_rate = rate

    @staticmethod
    def validate(amount): return amount > 0

    def __repr__(self): return f'BankAccount({self.owner}, {self._balance})'

acc = BankAccount('Musk', 1000)
acc.deposit(500)
print(acc.balance, acc)
```

```python
# --- Inheritance & super() ---
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0):
        super().__init__(owner, balance)
        self.locked = False

    def withdraw(self, amount):
        if self.locked:
            raise PermissionError('Account locked')
        super().withdraw(amount)

sa = SavingsAccount('Alice', 500)
sa.withdraw(200)
print(sa.balance)
```

```python
# ⚠️ Trap: class variable shared across instances
class Counter:
    count = 0
    def __init__(self):
        Counter.count += 1  # modify via class name, not self

Counter(); Counter()
print(Counter.count)  # 2

# Borderline: __slots__ prevents dynamic attributes
class Point:
    __slots__ = ['x','y']
    def __init__(self,x,y): self.x=x; self.y=y

p = Point(1,2)
try:
    p.z = 3  # AttributeError
except AttributeError as e:
    print(e)
```

## Chapter 16 — OOP With Polymorphism

```python
# --- Method overriding & duck typing ---
class Shape:
    def area(self): raise NotImplementedError

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14159 * self.r**2

class Rectangle(Shape):
    def __init__(self, w, h): self.w=w; self.h=h
    def area(self): return self.w * self.h

shapes = [Circle(5), Rectangle(4,6)]
for s in shapes:
    print(f'{s.__class__.__name__}: area={s.area():.2f}')
```

```python
# --- Dunder/magic methods (operator overloading) ---
class Vector:
    def __init__(self, x, y): self.x=x; self.y=y
    def __add__(self, other): return Vector(self.x+other.x, self.y+other.y)
    def __mul__(self, scalar): return Vector(self.x*scalar, self.y*scalar)
    def __repr__(self): return f'Vector({self.x}, {self.y})'

v1, v2 = Vector(1,2), Vector(3,4)
print(v1 + v2)    # Vector(4, 6)
print(v1 * 3)     # Vector(3, 6)
```

```python
# --- Abstract Base Class ---
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self): ...

class Dog(Animal):
    def speak(self): return 'Woof'

# ⚠️ Trap: instantiating abstract class raises TypeError
try:
    Animal()
except TypeError as e:
    print(e)

print(Dog().speak())
```

## Chapter 17 — Regular Expressions

```python
import re

text = 'Call us at 123-456-7890 or 098-765-4321'

# --- search / match / findall ---
print(re.search(r'\d{3}-\d{3}-\d{4}', text).group())
print(re.findall(r'\d{3}-\d{3}-\d{4}', text))
print(re.sub(r'\d', '*', text))  # mask digits
```

```python
# --- Named groups & compile ---
pattern = re.compile(r'(?P<area>\d{3})-(?P<num>\d{3}-\d{4})')
m = pattern.search(text)
if m:
    print(m.group('area'), m.group('num'))
```

```python
# ⚠️ Trap: re.match only checks at start of string
print(re.match(r'\d+', 'abc123'))  # None — not at start
print(re.search(r'\d+', 'abc123').group())  # '123'

# Borderline: greedy vs non-greedy
html = '<b>bold</b> and <i>italic</i>'
print(re.findall(r'<.+>',  html))   # greedy  — 1 big match
print(re.findall(r'<.+?>', html))   # lazy    — 4 tags
```

## Chapter 18 — Multi-Threading

```python
import threading, time

def worker(name, delay):
    time.sleep(delay)
    print(f'[{name}] done after {delay}s')

t1 = threading.Thread(target=worker, args=('A', 0.1))
t2 = threading.Thread(target=worker, args=('B', 0.05))

t1.start(); t2.start()
t1.join();  t2.join()
print('All threads finished')
```

```python
# --- Race condition & Lock ---
lock = threading.Lock()
counter = 0

def safe_increment():
    global counter
    with lock:
        counter += 1

threads = [threading.Thread(target=safe_increment) for _ in range(1000)]
for t in threads: t.start()
for t in threads: t.join()
print('Counter:', counter)  # always 1000
```

```python
# ⚠️ Trap: GIL limits CPU-bound thread parallelism
# Threads = good for I/O-bound tasks (network, disk)
# For CPU-bound → use multiprocessing.Pool

# Borderline: daemon threads die when main thread exits
def bg():
    while True: pass

d = threading.Thread(target=bg, daemon=True)
d.start()
print('Main done — daemon auto-killed')
```

## Chapter 19 — Python Database Programming

```python
import sqlite3

# --- Connect / Create ---
conn = sqlite3.connect(':memory:')  # in-memory DB
cur  = conn.cursor()

cur.execute('''
    CREATE TABLE students (
        id   INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        grade REAL
    )
''')
conn.commit()
```

```python
# --- Insert (parameterised — prevents SQL injection) ---
students = [('Alice', 92.5), ('Bob', 78.0), ('Carol', 88.5)]
cur.executemany('INSERT INTO students (name,grade) VALUES (?,?)', students)
conn.commit()

# --- Select ---
cur.execute('SELECT * FROM students WHERE grade > ?', (80,))
for row in cur.fetchall():
    print(row)

# --- Update / Delete ---
cur.execute('UPDATE students SET grade=95 WHERE name=?', ('Alice',))
cur.execute('DELETE FROM students WHERE name=?', ('Bob',))
conn.commit()
```

```python
# ⚠️ Trap: string formatting in SQL = injection risk
# BAD:  cur.execute(f'SELECT * FROM t WHERE name="{user_input}"')
# GOOD: cur.execute('SELECT * FROM t WHERE name=?', (user_input,))

# Borderline: connection not closed → data loss without commit
try:
    cur.execute('INSERT INTO students (name,grade) VALUES (?,?)', ('Dave',70))
    # if crash here without commit → row lost
    conn.commit()
except Exception:
    conn.rollback()
finally:
    conn.close()
print('DB session complete')
```
