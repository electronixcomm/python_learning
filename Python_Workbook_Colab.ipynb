```python
# ─── Float Special Values & Precision ───────────────────────────────────────
import sys

print(sys.float_info.max)     # largest float (~1.8e308)
print(sys.float_info.min)     # smallest positive float

inf  = float('inf')
ninf = float('-inf')
nan  = float('nan')
print(inf, ninf, nan)

print(inf > 1e308)    # True
print(nan == nan)     # False  ⚠️ TRAP: NaN is never equal to itself!
import math
print(math.isnan(nan))   # True  — correct way to check NaN

# Floating point representation trap
print(0.1 + 0.2)          # 0.30000000000000004  NOT 0.3!
print(0.1 + 0.2 == 0.3)   # False  ⚠️ TRAP: float arithmetic is not exact
print(abs(0.1 + 0.2 - 0.3) < 1e-9)  # True  — correct float comparison

# Use round() or decimal module for precision
from decimal import Decimal
print(Decimal("0.1") + Decimal("0.2"))  # 0.3  exact!
```

```python
# ─── Mutable vs Immutable ────────────────────────────────────────────────────
# 💡 Theory hint:
#   Immutable: int, float, complex, bool, str, tuple, bytes, frozenset
#   Mutable:   list, dict, set, bytearray

# Immutable — id changes on "modification"
x = 10
print(id(x))
x = x + 1    # creates a NEW int object; x now points to 11
print(id(x)) # different id

# Mutable — id stays same on modification
lst = [1, 2, 3]
print(id(lst))
lst.append(4)   # modifies in-place
print(id(lst))  # SAME id

# ⚠️ TRAP: strings are immutable — every operation returns a NEW string
s = "Hello"
s2 = s.upper()   # new string
print(s, s2, s is s2)   # Hello  HELLO  False

# ⚠️ TRAP: passing mutable objects to functions — changes affect the caller!
def add_item(lst, item):
    lst.append(item)   # mutates the original list

data = [1, 2, 3]
add_item(data, 99)
print(data)   # [1, 2, 3, 99] — original modified!

# Fix: pass a copy
add_item(data.copy(), 100)
print(data)   # [1, 2, 3, 99] — original unchanged
```

## Chapter 2 — Operators

```python
# ─── Arithmetic Operators ───────────────────────────────────────────────────
print(10 / 3)    # 3.3333…  — true division always returns float
print(10 // 3)   # 3        — floor division
print(10 % 3)    # 1        — modulo
print(2 ** 10)   # 1024     — exponentiation (right-associative)
print(-7 // 2)   # -4       — floor towards negative infinity
# ⚠️ TRAP: -7//2 is -4, NOT -3  (floor, not truncate)
print(-7 % 2)    # 1        — result has same sign as divisor
```

```python
# ─── Relational / Comparison ────────────────────────────────────────────────
print(1 < 2 < 3)     # True  — chained comparison (Pythonic)
print(1 < 3 > 2)     # True  — all pairs checked
# ⚠️ TRAP: In most languages chained comparisons don't work; Python is special
```

```python
# ─── Logical Operators & Short-Circuit ──────────────────────────────────────
# 'and' returns first falsy value or last value
print(0 and 5)    # 0
print(3 and 5)    # 5
print([] and 10)  # []

# 'or' returns first truthy value or last value
print(0 or 5)     # 5
print(3 or 5)     # 3
print([] or {})   # {}

# not
print(not 0)      # True
print(not [])     # True
# 🔲 BORDERLINE: 'not' returns bool; 'and'/'or' return operand values, not bool
```

```python
# ─── Bitwise Operators ──────────────────────────────────────────────────────
print(4 & 5)   # 4   — AND
print(4 | 5)   # 5   — OR
print(4 ^ 5)   # 1   — XOR
print(~4)      # -5  — bitwise NOT  (~x = -(x+1))
print(2 << 3)  # 16  — left shift  (×2 per shift)
print(16 >> 2) # 4   — right shift (÷2 per shift)
# ⚠️ TRAP: ~x = -(x+1), so ~0 = -1, ~-1 = 0
```

```python
# ─── Ternary / Conditional Expression ───────────────────────────────────────
x = 10
result = "Even" if x % 2 == 0 else "Odd"
print(result)   # Even

# Nested ternary (use sparingly)
grade = "A" if x >= 90 else ("B" if x >= 75 else "C")
print(grade)
```

```python
# ─── Identity and Membership ─────────────────────────────────────────────────
print(10 is 10)          # True (small int cache)
print([1,2] is [1,2])    # False — two different list objects

print(3 in [1,2,3])      # True
print('a' in 'apple')    # True
print(4 not in [1,2,3])  # True

# ─── Operator Precedence (high → low) ──────────────────────────────────────
# **  >  ~,+,-  >  *,/,//,%  >  +,-  >  <<,>>  >  &  >  ^  >  |  >  comparisons
# >  not  >  and  >  or
print(2 + 3 * 4)    # 14  (multiplication first)
print(2 ** 3 ** 2)  # 512 (right-associative: 3**2=9, 2**9=512)
```

```python
# ─── Augmented Assignment Operators ─────────────────────────────────────────
# 💡 These modify a variable in-place (same as x = x op value)
x = 10
x += 5;  print(x)   # 15
x -= 3;  print(x)   # 12
x *= 2;  print(x)   # 24
x //= 5; print(x)   # 4
x **= 3; print(x)   # 64
x %= 10; print(x)   # 4
x /= 2;  print(x)   # 2.0  (always float)

# Bitwise augmented
a = 0b1010   # 10
a &= 0b1100  # AND in-place → 0b1000 = 8
print(bin(a))

a |= 0b0011  # OR in-place → 0b1011 = 11
print(bin(a))

# ⚠️ TRAP: augmented assignment on immutable types still creates a NEW object
s = "Hello"
old_id = id(s)
s += " World"    # creates new string object
print(id(s) == old_id)  # False — new object

# ⚠️ TRAP: augmented assignment on mutable types modifies IN-PLACE
lst = [1, 2, 3]
old_id = id(lst)
lst += [4, 5]    # same as lst.extend([4,5]) — modifies in-place
print(id(lst) == old_id)  # True — same object!
```

```python
# ─── divmod() — quotient and remainder together ─────────────────────────────
# Custom example: useful for time/unit conversions

q, r = divmod(17, 5)
print(f"17 ÷ 5 = {q} remainder {r}")  # 3 remainder 2

# ✅ Custom: Convert seconds to hours, minutes, seconds
total_seconds = 3725
hours,   rem  = divmod(total_seconds, 3600)
minutes, secs = divmod(rem, 60)
print(f"{total_seconds}s = {hours}h {minutes}m {secs}s")   # 1h 2m 5s

# ✅ Custom: Check divisibility
def is_divisible(n, d):
    _, r = divmod(n, d)
    return r == 0

print(is_divisible(100, 4))   # True
print(is_divisible(100, 7))   # False

# 🔲 BORDERLINE: divmod with floats
print(divmod(7.5, 2.5))   # (3.0, 0.0) — works, returns floats
print(divmod(-7, 2))       # (-4, 1)    — floor division rules apply
```

```python
# ─── Bitwise Operators — Practical Tricks ──────────────────────────────────
# 💡 Theory hint: Bitwise ops work on the binary representation of integers

# ✅ Custom Trick 1: Check if a number is even or odd (faster than %)
def is_even(n): return (n & 1) == 0
def is_odd(n):  return (n & 1) == 1

print(is_even(10), is_odd(7))   # True True

# ✅ Custom Trick 2: Swap two numbers without temp variable using XOR
a, b = 5, 9
a ^= b
b ^= a
a ^= b
print(a, b)   # 9 5 — swapped!

# ✅ Custom Trick 3: Multiply / Divide by powers of 2 using shifts
x = 8
print(x << 1)   # 16  (×2)
print(x << 2)   # 32  (×4)
print(x >> 1)   # 4   (÷2)
print(x >> 3)   # 1   (÷8)

# ✅ Custom Trick 4: Bit flags (used in permissions, feature toggles)
READ    = 0b001   # 1
WRITE   = 0b010   # 2
EXECUTE = 0b100   # 4

perms = READ | WRITE     # grant read+write
print(bin(perms))        # 0b11

print(bool(perms & READ))     # True  — has read?
print(bool(perms & EXECUTE))  # False — has execute?

perms |= EXECUTE             # grant execute
print(bool(perms & EXECUTE))  # True

perms &= ~WRITE              # revoke write
print(bool(perms & WRITE))    # False

# ⚠️ TRAP: ~ (bitwise NOT) on n gives -(n+1), not the expected bit-flip in isolation
print(~0b1010)   # -11  not 0b0101
```

```python
# ─── Walrus Operator := (Python 3.8+) ───────────────────────────────────────
# 💡 Theory hint: := assigns AND returns a value in a single expression
# Useful to avoid calling the same expression twice

import re

# ✅ Custom Example 1: while loop that reads and checks in one step
data = [15, 22, 8, 31, 4, 19]
i = 0
while (val := data[i]) < 25:   # assign val AND test it
    print(f"Under 25: {val}")
    i += 1

# ✅ Custom Example 2: avoid double computation in if
import math
numbers = [16, -4, 25, 0, 9]
for n in numbers:
    if (root := math.sqrt(abs(n))) > 3:
        print(f"sqrt(|{n}|) = {root:.2f} > 3")

# ✅ Custom Example 3: list comprehension with reuse
results = [y for x in range(10) if (y := x**2) > 20]
print(results)   # [25, 36, 49, 64, 81]

# ⚠️ TRAP: walrus operator needs parentheses in most contexts
# if x := 10:  is valid but can look confusing — always use parens for clarity
```

```python
# ─── Short-Circuit Evaluation — Practical Patterns ──────────────────────────
# 💡 Theory hint: Python stops evaluating as soon as result is determined
#   and  → stops at first Falsy
#   or   → stops at first Truthy

# ✅ Custom Pattern 1: safe default value
user_input = ""
name = user_input or "Anonymous"
print(name)   # Anonymous

config = {"timeout": 0}
timeout = config.get("timeout") or 30   # ⚠️ TRAP: 0 is falsy! Gets 30 instead of 0
print(timeout)   # 30  (probably wrong)
# Better: use explicit None check
timeout = config.get("timeout") if config.get("timeout") is not None else 30
print(timeout)   # 0  (correct)

# ✅ Custom Pattern 2: guard before expensive/risky call
def fetch_data():
    print("Fetching...")
    return [1, 2, 3]

cache = [1, 2, 3]
result = cache or fetch_data()   # fetch_data() never called — short-circuit!
print(result)

# ✅ Custom Pattern 3: inline validation
def safe_divide(a, b):
    return b != 0 and a / b   # returns False if b==0, else the result

print(safe_divide(10, 2))    # 5.0
print(safe_divide(10, 0))    # False

# 🔲 BORDERLINE: 'and'/'or' don't return True/False, they return the actual operand
print(type("hello" or "world"))   # str, not bool
print(type(not "hello"))          # bool — 'not' always returns bool
```

```python
# ─── Operator Precedence — Full Working Examples ────────────────────────────
# 💡 Precedence (highest first):
#   ()  →  **  →  +x,-x,~x  →  *,/,//,%  →  +,-
#   →  <<,>>  →  &  →  ^  →  |  →  comparisons  →  not  →  and  →  or

# ✅ Custom: predict before running each line
print(2 + 3 * 4)          # 14   not 20  (* before +)
print(2 ** 3 ** 2)        # 512  not 64  (right-associative: 3**2=9, 2**9=512)
print(-2 ** 2)            # -4   not 4   (** before unary -)
print((-2) ** 2)          # 4    parens first
print(True + True * 3)    # 4    (True=1; 1*3=3; 1+3=4)
print(not 1 + 1 == 2)     # False  (1+1=2, 2==2→True, not True→False)
print(3 > 2 > 1)          # True  (chained: 3>2 AND 2>1)
print(3 > 2 and 2 > 1)    # True  (same meaning, explicit)

# ⚠️ TRAP: comparison operators are evaluated TOGETHER (same precedence, left-right)
# but 'not', 'and', 'or' have LOWER precedence than comparisons
print(not 3 > 2)          # False  (3>2=True, not True=False)
print(1 == 1 or 0 and 0)  # True   (and before or: 0 and 0=0, 1==1 or 0 = True)
```

```python
# ─── match-case (Structural Pattern Matching, Python 3.10+) ─────────────────
# 💡 Theory hint: Python's answer to switch/case from other languages
# Much more powerful — matches structures, not just values

command = "quit"
match command:
    case "go north" | "go south":
        print("Moving...")
    case "quit" | "exit":
        print("Goodbye!")
    case _:                     # default case (wildcard)
        print("Unknown command")

# ✅ Custom: match with capture
point = (0, 5)
match point:
    case (0, 0):  print("Origin")
    case (x, 0):  print(f"On X-axis at {x}")
    case (0, y):  print(f"On Y-axis at {y}")
    case (x, y):  print(f"Point at ({x}, {y})")
```

```python
# ─── Loop patterns — enumerate, zip, range tricks ───────────────────────────
# ✅ Custom: enumerate — get index + value together
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits, start=1):    # start index at 1
    print(f"{i}. {fruit}")

# ✅ Custom: zip — iterate multiple lists together
names   = ["Alice", "Bob", "Charlie"]
scores  = [85, 92, 78]
grades  = ["B", "A", "C"]
for name, score, grade in zip(names, scores, grades):
    print(f"{name}: {score} ({grade})")

# ⚠️ TRAP: zip stops at the SHORTEST iterable
print(list(zip([1,2,3], ["a","b"])))   # [(1,'a'),(2,'b')] — 3 is dropped!

# Use zip_longest for full coverage
from itertools import zip_longest
print(list(zip_longest([1,2,3], ["a","b"], fillvalue="?")))
```

```python
# ─── Nested loops — practical patterns ──────────────────────────────────────
# ✅ Custom: find all pairs that sum to a target
nums = [1, 4, 3, 7, 2, 6]
target = 7
pairs = [(nums[i], nums[j])
         for i in range(len(nums))
         for j in range(i+1, len(nums))
         if nums[i] + nums[j] == target]
print("Pairs summing to", target, ":", pairs)

# ✅ Custom: breaking out of nested loops using flag
found = False
for i in range(5):
    for j in range(5):
        if i * j == 12:
            print(f"Found: {i} × {j} = 12")
            found = True
            break
    if found:
        break

# ✅ Custom: Pythonic alternative — use a function and return
def find_product(target, n):
    for i in range(n):
        for j in range(n):
            if i * j == target:
                return i, j
    return None

print(find_product(12, 5))   # (3, 4)
```

```python
# ─── List — stack and queue usage ───────────────────────────────────────────
# 💡 Theory hint: lists work well as stacks; use deque for efficient queues

# Stack (LIFO) — append/pop from end
stack = []
stack.append("task1")
stack.append("task2")
stack.append("task3")
print("Pop:", stack.pop())   # task3
print("Stack:", stack)

# Queue using collections.deque (efficient O(1) popleft)
from collections import deque
queue = deque()
queue.append("A")
queue.append("B")
queue.append("C")
print("Dequeue:", queue.popleft())  # A
print("Queue:", queue)

# ⚠️ TRAP: list.pop(0) is O(n) — shifts all elements; deque.popleft() is O(1)

# ✅ Custom: browser history (stack)
history = []
def visit(url):
    history.append(url)
    print(f"Visiting: {url}")

def back():
    if len(history) > 1:
        history.pop()
        print(f"Back to: {history[-1]}")
    else:
        print("No history")

visit("google.com")
visit("github.com")
visit("stackoverflow.com")
back()
back()
```

```python
# ─── Tuple — named tuples ────────────────────────────────────────────────────
# 💡 Theory hint: namedtuple gives tuples field names — more readable than index
from collections import namedtuple

Point  = namedtuple('Point', ['x', 'y'])
Person = namedtuple('Person', ['name', 'age', 'city'])

p = Point(3, 4)
print(p.x, p.y)        # 3 4
print(p[0], p[1])      # 3 4  (also works by index)
print(p)               # Point(x=3, y=4)

# Distance formula
import math
def distance(p1, p2):
    return math.sqrt((p2.x-p1.x)**2 + (p2.y-p1.y)**2)

print(distance(Point(0,0), Point(3,4)))  # 5.0

emp = Person("Durga", 40, "Hyderabad")
print(emp.name, emp.age)    # Durga 40
print(emp._asdict())        # OrderedDict — convert to dict
print(emp._replace(age=41)) # create modified copy

# ⚠️ TRAP: namedtuples are IMMUTABLE like regular tuples
# emp.age = 41   # AttributeError
```

```python
# ─── Set — real-world use cases ─────────────────────────────────────────────
# ✅ Custom: remove duplicates while preserving order
def unique_ordered(seq):
    seen = set()
    return [x for x in seq if not (x in seen or seen.add(x))]

data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
print(unique_ordered(data))   # [3, 1, 4, 5, 9, 2, 6]

# ✅ Custom: find common elements between two lists (fast with sets)
list1 = [1, 2, 3, 4, 5, 6]
list2 = [4, 5, 6, 7, 8, 9]
common = set(list1) & set(list2)
only_in_1 = set(list1) - set(list2)
only_in_2 = set(list2) - set(list1)
print("Common  :", common)      # {4, 5, 6}
print("Only L1 :", only_in_1)   # {1, 2, 3}
print("Only L2 :", only_in_2)   # {7, 8, 9}

# ✅ Custom: membership test — set is O(1), list is O(n)
import time
big_list = list(range(1_000_000))
big_set  = set(big_list)

start = time.time(); _ = 999999 in big_list; print(f"List search: {time.time()-start:.6f}s")
start = time.time(); _ = 999999 in big_set;  print(f"Set  search: {time.time()-start:.6f}s")
# Set is dramatically faster for 'in' checks
```

```python
# ─── Decorators ──────────────────────────────────────────────────────────────
# 💡 Theory hint: A decorator is a function that takes a function and returns
#    an enhanced function. The @syntax is syntactic sugar.

def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
# Output: Before  Hello!  After

# ─── Decorator Chaining ───────────────────────────────────────────────────────
def bold(func):
    def wrapper(): return "<b>" + func() + "</b>"
    return wrapper

def italic(func):
    def wrapper(): return "<i>" + func() + "</i>"
    return wrapper

@bold
@italic
def text():
    return "Hello"

print(text())   # <b><i>Hello</i></b>  — applied bottom-up
```

```python
# ─── Generators (yield) ──────────────────────────────────────────────────────
# 💡 Theory hint: Generators are lazy — they compute values one at a time
# Much more memory-efficient than building a full list

def even_gen(limit):
    for i in range(0, limit+1, 2):
        yield i

gen = even_gen(10)
print(next(gen))   # 0
print(next(gen))   # 2
print(list(gen))   # [4, 6, 8, 10]  — remaining values

# Infinite generator
def count_up(start=0):
    while True:
        yield start
        start += 1

counter = count_up(1)
print([next(counter) for _ in range(5)])   # [1, 2, 3, 4, 5]

# ⚠️ TRAP: once a generator is exhausted it raises StopIteration
gen2 = even_gen(4)
print(list(gen2))  # [0, 2, 4]
print(list(gen2))  # []  — exhausted!
```

```python
# ─── Function annotations (type hints) ──────────────────────────────────────
# 💡 Theory hint: type hints are NOT enforced at runtime — only documentation
# Use mypy or pyright for static type checking

def greet(name: str, times: int = 1) -> str:
    return (f"Hello, {name}! " * times).strip()

print(greet("Durga", 3))
print(greet.__annotations__)   # {'name': <class 'str'>, 'times': <class 'int'>, 'return': <class 'str'>}

# ✅ Custom: more complex annotations
from typing import List, Dict, Optional, Union, Tuple

def process(data: List[int]) -> Dict[str, float]:
    return {"min": min(data), "max": max(data), "avg": sum(data)/len(data)}

print(process([4, 8, 15, 16, 23, 42]))

# Optional — value can be that type OR None
def find_user(uid: int) -> Optional[str]:
    users = {1:"Alice", 2:"Bob"}
    return users.get(uid)

print(find_user(1))    # Alice
print(find_user(99))   # None

# Union — multiple types allowed
def double(x: Union[int, float]) -> Union[int, float]:
    return x * 2

print(double(5), double(3.14))
```

```python
# ─── Closures and nonlocal ───────────────────────────────────────────────────
# 💡 Theory hint: a closure is a function that remembers variables from its
# enclosing scope even after that scope has finished executing

def make_counter(start=0):
    count = start
    def counter():
        nonlocal count       # access enclosing scope variable
        count += 1
        return count
    return counter

c1 = make_counter()
c2 = make_counter(10)
print(c1(), c1(), c1())    # 1 2 3
print(c2(), c2())           # 11 12  — independent from c1

# ✅ Custom: multiplier factory
def make_multiplier(n):
    return lambda x: x * n    # lambda closes over n

double = make_multiplier(2)
triple = make_multiplier(3)
print(double(5), triple(5))   # 10 15

# ⚠️ TRAP: late binding in closures inside loops
funcs = [lambda: i for i in range(3)]   # WRONG
print([f() for f in funcs])              # [2,2,2] — all see final i=2!

funcs_fixed = [lambda i=i: i for i in range(3)]   # capture i by default arg
print([f() for f in funcs_fixed])                  # [0,1,2] — correct
```

```python
# ─── Decorators — practical examples ────────────────────────────────────────
import time, functools

# ✅ Custom: timer decorator
def timer(func):
    @functools.wraps(func)     # preserves original function metadata
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        print(f"{func.__name__} took {end-start:.6f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(range(n))

print(slow_sum(1_000_000))

# ✅ Custom: retry decorator
def retry(times=3):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(1, times+1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"Attempt {attempt} failed: {e}")
            raise RuntimeError(f"{func.__name__} failed after {times} attempts")
        return wrapper
    return decorator

import random
@retry(times=3)
def flaky_operation():
    if random.random() < 0.7:   # fails 70% of the time
        raise ValueError("Random failure!")
    return "Success!"

try:
    print(flaky_operation())
except RuntimeError as e:
    print(e)
```

```python
# ─── Generators — advanced patterns ─────────────────────────────────────────
# ✅ Custom: infinite sequence generator
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
print([next(fib) for _ in range(10)])   # first 10 Fibonacci numbers

# ✅ Custom: data pipeline with generators (memory efficient)
def read_numbers(n):        # simulate reading from large file
    for i in range(1, n+1):
        yield i

def filter_even(nums):
    return (n for n in nums if n % 2 == 0)

def square(nums):
    return (n*n for n in nums)

def take(n, gen):
    for _ in range(n):
        yield next(gen)

# Pipeline: read → filter evens → square → take first 5
pipeline = take(5, square(filter_even(read_numbers(100))))
print(list(pipeline))    # [4, 16, 36, 64, 100]

# ✅ Custom: generator send() — two-way communication
def accumulator():
    total = 0
    while True:
        value = yield total   # yield sends total OUT; receives value IN
        if value is None: break
        total += value

acc = accumulator()
next(acc)            # prime the generator
print(acc.send(10))  # 10
print(acc.send(20))  # 30
print(acc.send(5))   # 35
```

```python
# --- Custom exception hierarchy ---
# ✅ Custom: build a bank account with rich custom exceptions

class BankError(Exception):
    '''Base class for all bank exceptions'''
    pass

class InsufficientFundsError(BankError):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount  = amount
        super().__init__(
            f"Cannot withdraw {amount}. Balance is {balance}."
        )

class NegativeAmountError(BankError):
    def __init__(self, amount):
        super().__init__(f"Amount must be positive, got {amount}")

class AccountFrozenError(BankError):
    pass

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner   = owner
        self._balance = balance
        self._frozen = False

    def deposit(self, amount):
        if amount <= 0:
            raise NegativeAmountError(amount)
        self._balance += amount
        return self._balance

    def withdraw(self, amount):
        if self._frozen:
            raise AccountFrozenError("Account is frozen")
        if amount <= 0:
            raise NegativeAmountError(amount)
        if amount > self._balance:
            raise InsufficientFundsError(self._balance, amount)
        self._balance -= amount
        return self._balance

    def freeze(self): self._frozen = True

acc = BankAccount("Durga", 1000)
acc.deposit(500)
print("Balance:", acc.withdraw(200))   # 1300

for action in [
    lambda: acc.withdraw(5000),
    lambda: acc.deposit(-100),
]:
    try:
        action()
    except BankError as e:   # catches ALL bank-related exceptions
        print(f"{type(e).__name__}: {e}")
```

```python
# ─── Class, Object, __init__, self ──────────────────────────────────────────
class Student:
    '''Student class — tracks name, rollno, marks'''

    def __init__(self, name, rollno, marks):
        # instance variables — unique per object
        self.name = name
        self.rollno = rollno
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}  Roll: {self.rollno}  Marks: {self.marks}")

    def grade(self):
        if self.marks >= 60:   return "First Class"
        elif self.marks >= 50: return "Second Class"
        elif self.marks >= 35: return "Third Class"
        else:                   return "Fail"

s1 = Student("Durga", 101, 85)
s2 = Student("Ravi",  102, 45)
s1.display()
print(s1.grade())
s2.display()
print(s2.grade())

print(Student.__doc__)
```

```python
# ─── Static Variables (Class Variables) ──────────────────────────────────────
class Test:
    count = 0              # class variable — shared by all objects

    def __init__(self):
        Test.count += 1    # modify via class name

    @classmethod
    def get_count(cls):
        return cls.count

t1, t2, t3 = Test(), Test(), Test()
print(Test.get_count())    # 3

# ⚠️ TRAP: modifying class variable via self creates an INSTANCE variable instead
t1.count = 999             # creates t1.count instance var; Test.count unchanged
print(Test.count)          # 3  — class var still 3
print(t1.count)            # 999 — instance var
```

```python
# ─── Instance / Class / Static Methods ───────────────────────────────────────
class MathOps:
    pi = 3.14159

    def __init__(self, value):
        self.value = value

    def double(self):           # instance method — has self
        return self.value * 2

    @classmethod
    def circle_area(cls, r):    # class method — has cls
        return cls.pi * r * r

    @staticmethod
    def add(a, b):              # static method — no self/cls
        return a + b

m = MathOps(5)
print(m.double())              # 10
print(MathOps.circle_area(3))  # 28.27…
print(MathOps.add(10, 20))     # 30
```

```python
# ─── Inheritance ─────────────────────────────────────────────────────────────
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Employee(Person):
    def __init__(self, name, age, eno, esal):
        super().__init__(name, age)   # call parent constructor
        self.eno  = eno
        self.esal = esal

    def info(self):              # override parent method
        super().info()           # still call parent
        print(f"EmpNo: {self.eno}, Salary: {self.esal}")

e = Employee("Durga", 40, 101, 50000)
e.info()

print(isinstance(e, Employee))  # True
print(isinstance(e, Person))    # True  — IS-A relationship
```

```python
# ─── Dunder (Magic) Methods ──────────────────────────────────────────────────
class Vector:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __str__(self):           # print() calls this
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):          # repr() / interactive shell
        return f"Vector({self.x!r}, {self.y!r})"

    def __add__(self, other):    # v1 + v2
        return Vector(self.x + other.x, self.y + other.y)

    def __len__(self):           # len(v)
        return 2

    def __eq__(self, other):     # v1 == v2
        return self.x == other.x and self.y == other.y

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1)           # Vector(1, 2)
print(v1 + v2)      # Vector(4, 6)
print(len(v1))      # 2
print(v1 == Vector(1, 2))  # True
```

```python
# --- Dunder methods: complete practical set ---
# ✅ Custom: a fully-featured Fraction class

class Fraction:
    from math import gcd as _gcd

    def __init__(self, num, den=1):
        if den == 0:
            raise ZeroDivisionError("Denominator cannot be zero")
        g = Fraction._gcd(abs(num), abs(den))
        sign = -1 if den < 0 else 1
        self.num = sign * num // g
        self.den = sign * den // g

    def __repr__(self):  return f"Fraction({self.num},{self.den})"
    def __str__(self):   return f"{self.num}/{self.den}" if self.den!=1 else str(self.num)
    def __add__(self, o):return Fraction(self.num*o.den + o.num*self.den, self.den*o.den)
    def __sub__(self, o):return Fraction(self.num*o.den - o.num*self.den, self.den*o.den)
    def __mul__(self, o):return Fraction(self.num*o.num, self.den*o.den)
    def __truediv__(self,o):return Fraction(self.num*o.den, self.den*o.num)
    def __eq__(self, o): return self.num==o.num and self.den==o.den
    def __lt__(self, o): return self.num*o.den < o.num*self.den
    def __float__(self): return self.num / self.den
    def __bool__(self):  return self.num != 0

a = Fraction(1, 2)
b = Fraction(1, 3)
print(f"{a} + {b} = {a+b}")    # 1/2 + 1/3 = 5/6
print(f"{a} * {b} = {a*b}")    # 1/2 * 1/3 = 1/6
print(f"{a} / {b} = {a/b}")    # 1/2 / 1/3 = 3/2
print(a == Fraction(2,4))       # True (reduced)
print(a > b)                    # True
print(float(a))                 # 0.5
print(bool(Fraction(0,5)))      # False
```

```python
# --- Abstract Base Classes (ABC) ---
# 💡 Theory hint: ABC enforces method contracts -- subclasses MUST implement abstract methods
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self) -> float:
        pass

    @abstractmethod
    def perimeter(self) -> float:
        pass

    def describe(self):           # concrete method -- inherited as-is
        print(f"{type(self).__name__}: area={self.area():.2f}, perimeter={self.perimeter():.2f}")

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self):      return 3.14159 * self.r ** 2
    def perimeter(self): return 2 * 3.14159 * self.r

class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self):      return self.w * self.h
    def perimeter(self): return 2 * (self.w + self.h)

shapes = [Circle(5), Rectangle(4, 6), Circle(3)]
for s in shapes:
    s.describe()                  # polymorphic call

# ⚠️ TRAP: You cannot instantiate an ABC directly
try:
    Shape()
except TypeError as e:
    print(f"Can't instantiate ABC: {e}")

# 🔲 BORDERLINE: ABC vs Protocol -- ABC uses inheritance; Protocol uses structural subtyping
```

```python
# --- Duck Typing + Protocol (Python 3.8+) ---
# 💡 Theory hint: "If it quacks like a duck, it IS a duck" -- no inheritance needed
from typing import Protocol, runtime_checkable

@runtime_checkable
class Drawable(Protocol):
    def draw(self) -> None: ...
    def resize(self, factor: float) -> None: ...

class SVGCircle:
    def draw(self): print("Drawing SVG circle")
    def resize(self, f): print(f"Resizing circle by {f}x")

class CanvasRect:
    def draw(self): print("Drawing canvas rect")
    def resize(self, f): print(f"Resizing rect by {f}x")

class PlainText:            # does NOT implement Drawable
    def show(self): print("Showing text")

def render_all(items):
    for item in items:
        if isinstance(item, Drawable):  # runtime check via @runtime_checkable
            item.draw()
            item.resize(2.0)
        else:
            print(f"{type(item).__name__} is not Drawable -- skipping")

render_all([SVGCircle(), CanvasRect(), PlainText()])

# ✅ Duck typing without Protocol
def make_sound(animal):
    animal.speak()                # works for ANY object with speak()

class Dog:
    def speak(self): print("Woof!")
class Robot:
    def speak(self): print("Beep boop.")

for a in [Dog(), Robot()]:
    make_sound(a)
```

```python
# --- Method overriding + super() in polymorphism ---
# ✅ Custom: payment system with polymorphic process()

class Payment:
    def __init__(self, amount):
        self.amount = amount

    def validate(self):
        if self.amount <= 0:
            raise ValueError(f"Invalid amount: {self.amount}")

    def process(self):
        self.validate()
        print(f"Processing {type(self).__name__} for ${self.amount:.2f}")

class CreditCard(Payment):
    def __init__(self, amount, card_no):
        super().__init__(amount)
        self.card_no = card_no[-4:]     # store only last 4

    def process(self):
        super().process()               # run base validate + log
        print(f"  Charging card ending {self.card_no}")

class CryptoPayment(Payment):
    def __init__(self, amount, wallet):
        super().__init__(amount)
        self.wallet = wallet

    def validate(self):                 # override validation
        super().validate()
        if self.amount > 10000:
            raise ValueError("Crypto: single tx limit $10,000")

    def process(self):
        super().process()
        print(f"  Sending to wallet {self.wallet[:8]}...")

payments = [
    CreditCard(49.99, "4111111111111234"),
    CryptoPayment(500, "bc1qxy2kgdygjrsqtzq2n0yrf"),
    CreditCard(0, "4111111111115678"),
]

for p in payments:
    try:
        p.process()
    except ValueError as e:
        print(f"  ERROR: {e}")
```

```python
# --- Regex fundamentals + groups ---
# 💡 Theory hint: use raw strings r"..." to avoid double-escaping backslashes
import re

text = "Alice (age 30) and Bob (age 25) met at 3:45 PM on 2024-03-15"

# Named groups -- cleaner than positional
person_pat = re.compile(r'(?P<name>[A-Z][a-z]+) \(age (?P<age>\d+)\)')
for m in person_pat.finditer(text):
    print(f"Name: {m.group('name')}, Age: {m.group('age')}")

# Date extraction
date_pat = re.compile(r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})')
m = date_pat.search(text)
if m:
    print(f"Date: {m.group('year')}/{m.group('month')}/{m.group('day')}")

# Time extraction
time_m = re.search(r'(\d+):(\d{2}) (AM|PM)', text)
print(f"Time groups: {time_m.groups()}")

# ⚠️ TRAP: group(0) is the entire match; group(1) is the first capture group
# ⚠️ TRAP: findall() with groups returns list of tuples, not Match objects
names = re.findall(r'([A-Z][a-z]+) \(age (\d+)\)', text)
print(names)  # [('Alice', '30'), ('Bob', '25')]
```

```python
# --- Lookahead, lookbehind, non-capturing groups ---
# 💡 Theory hint: lookahead/behind are ZERO-WIDTH -- they assert but don't consume
import re

prices = "apple: $5.99  banana: $1.50  cherry: EUR12.00  mango: $8.75"

# Positive lookahead -- match digits followed by .
# Negative lookbehind -- NOT preceded by EUR
usd_pat = re.compile(r'(?<!EUR)\d+\.\d{2}')
print("USD amounts:", usd_pat.findall(prices))

# Positive lookbehind -- only digits preceded by $
dollar_pat = re.compile(r'(?<=\$)\d+\.\d{2}')
print("After $:", dollar_pat.findall(prices))

# Non-capturing group (?:...) -- group but don't capture
log = "ERROR: disk full | WARNING: low mem | INFO: started"
levels = re.findall(r'(?:ERROR|WARNING|INFO): ([^|]+)', log)
print("Messages:", [m.strip() for m in levels])

# 🔲 BORDERLINE: greedy vs lazy
html = "<b>bold</b> and <i>italic</i>"
greedy = re.findall(r'<.+>', html)      # entire string (greedy)
lazy   = re.findall(r'<.+?>', html)     # each tag separately (lazy)
print("Greedy:", greedy)
print("Lazy:  ", lazy)
```

```python
# --- Compiled patterns + validation library pattern ---
# ✅ Custom: reusable validator using compiled regex

import re
from dataclasses import dataclass
from typing import Optional

@dataclass
class ValidationResult:
    valid: bool
    message: str = ""

class Validator:
    # Compiled once at class level -- much faster in loops
    _EMAIL   = re.compile(r'^[\w.+-]+@[\w-]+\.[a-z]{2,}$', re.IGNORECASE)
    _PHONE   = re.compile(r'^\+?1?[-. ]?\(?\d{3}\)?[-. ]?\d{3}[-. ]?\d{4}$')
    _IP4     = re.compile(r'^(\d{1,3}\.){3}\d{1,3}$')
    _STRONG_PW = re.compile(
        r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!@#$%^&*]).{8,}$'
    )

    @classmethod
    def email(cls, s: str) -> ValidationResult:
        ok = bool(cls._EMAIL.match(s))
        return ValidationResult(ok, "" if ok else "Invalid email format")

    @classmethod
    def phone(cls, s: str) -> ValidationResult:
        ok = bool(cls._PHONE.match(s))
        return ValidationResult(ok, "" if ok else "Invalid phone number")

    @classmethod
    def ipv4(cls, s: str) -> ValidationResult:
        if not cls._IP4.match(s): return ValidationResult(False, "Bad format")
        parts = [int(p) for p in s.split(".")]
        ok = all(0 <= p <= 255 for p in parts)
        return ValidationResult(ok, "" if ok else "Octet out of range 0-255")

    @classmethod
    def strong_password(cls, s: str) -> ValidationResult:
        ok = bool(cls._STRONG_PW.match(s))
        msg = "" if ok else "Need 8+ chars, upper, lower, digit, special char"
        return ValidationResult(ok, msg)

tests = {
    "email":    ["alice@example.com", "bad@", "user.name+tag@domain.co.uk"],
    "phone":    ["+1 (555) 123-4567", "555-1234", "12345678901"],
    "ipv4":     ["192.168.1.1", "256.0.0.1", "10.0.0"],
    "password": ["Weak1", "Str0ng!Pass", "nouppercase1!"],
}

for check, values in tests.items():
    fn = getattr(Validator, check)
    for v in values:
        r = fn(v)
        status = "OK" if r.valid else f"FAIL: {r.message}"
        print(f"  {check:10} {v:30} -> {status}")
    print()
```

```python
from threading import Thread, current_thread
import time

print("Main thread:", current_thread().getName())

# ─── Method 1: Function-based thread ──────────────────────────────────────────
def display(name, n):
    for i in range(n):
        print(f"[{name}] count={i}")
        time.sleep(0.1)

t1 = Thread(target=display, args=("ChildThread", 3))
t1.start()
t1.join()   # wait for t1 to finish
print("Main thread continues")
```

```python
# ─── Performance: with vs without multithreading ─────────────────────────────
import time
from threading import Thread

def task(name, n):
    for i in range(n):
        time.sleep(0.05)

# Sequential
start = time.time()
task("A", 5)
task("B", 5)
print("Sequential:", round(time.time()-start, 2), "s")

# Concurrent
start = time.time()
t1 = Thread(target=task, args=("A", 5))
t2 = Thread(target=task, args=("B", 5))
t1.start(); t2.start()
t1.join();  t2.join()
print("Concurrent:", round(time.time()-start, 2), "s")
```

```python
# ─── Synchronization with Lock ───────────────────────────────────────────────
from threading import Lock

lock = Lock()
counter = 0

def increment():
    global counter
    for _ in range(100000):
        with lock:          # acquire and auto-release
            counter += 1

threads = [Thread(target=increment) for _ in range(5)]
for t in threads: t.start()
for t in threads: t.join()

print("Expected:", 500000)
print("Got:     ", counter)
# Without lock: counter would be < 500000 (race condition)
```

```python
# ─── Daemon Threads ──────────────────────────────────────────────────────────
# 💡 Theory hint: Daemon threads run in background.
# When ALL non-daemon threads finish, Python auto-kills daemon threads.

def background_task():
    i = 0
    while True:
        print(f"[Daemon] tick {i}")
        i += 1
        time.sleep(1)

d = Thread(target=background_task, daemon=True, name="DaemonThread")
d.start()
time.sleep(2.5)
print("Main thread done — daemon will be killed automatically")
```

```python
# ─── Inter-thread Communication (Queue) ──────────────────────────────────────
import queue, random

q = queue.Queue()

def producer(q):
    for _ in range(5):
        item = random.randint(1, 100)
        q.put(item)
        print(f"Produced: {item}")
        time.sleep(0.2)

def consumer(q):
    for _ in range(5):
        item = q.get()
        print(f"Consumed: {item}")
        q.task_done()

t1 = Thread(target=producer, args=(q,))
t2 = Thread(target=consumer, args=(q,))
t1.start(); t2.start()
t1.join();  t2.join()
print("Queue demo done")
```

```python
# --- Thread safety: race condition demo + fix ---
# 💡 Theory hint: threads share memory -- concurrent writes cause data races
import threading
import time

class UnsafeCounter:
    def __init__(self): self.value = 0
    def increment(self):
        v = self.value
        time.sleep(0.0001)      # simulate tiny delay
        self.value = v + 1

class SafeCounter:
    def __init__(self):
        self.value = 0
        self._lock = threading.Lock()
    def increment(self):
        with self._lock:        # atomic: only one thread at a time
            self.value += 1

def run_threads(counter, n=100):
    threads = [threading.Thread(target=counter.increment) for _ in range(n)]
    for t in threads: t.start()
    for t in threads: t.join()
    return counter.value

# Run multiple times to show non-determinism of unsafe counter
unsafe_results = [run_threads(UnsafeCounter()) for _ in range(3)]
safe_results   = [run_threads(SafeCounter())   for _ in range(3)]

print("Unsafe (expected 100 each):", unsafe_results)   # likely NOT 100
print("Safe   (expected 100 each):", safe_results)     # always 100

# ⚠️ TRAP: += is NOT atomic in Python -- it's read-modify-write
# ⚠️ TRAP: Lock is NOT reentrant; use RLock if same thread needs to acquire twice
```

```python
# --- ThreadPoolExecutor: the modern way ---
# 💡 Theory hint: ThreadPoolExecutor manages a pool of worker threads automatically
from concurrent.futures import ThreadPoolExecutor, as_completed
import time, random

def fetch_data(url_id):
    delay = random.uniform(0.1, 0.5)
    time.sleep(delay)               # simulate network call
    if url_id == 3:
        raise ConnectionError(f"Timeout on item {url_id}")
    return f"data_from_{url_id}"

print("=== as_completed (results as they finish) ===")
with ThreadPoolExecutor(max_workers=4) as pool:
    futures = {pool.submit(fetch_data, i): i for i in range(1, 7)}
    for fut in as_completed(futures):
        item = futures[fut]
        try:
            result = fut.result()
            print(f"  [{item}] => {result}")
        except Exception as e:
            print(f"  [{item}] ERROR: {e}")

print()
print("=== map (preserves order, simpler) ===")
with ThreadPoolExecutor(max_workers=4) as pool:
    results = pool.map(lambda i: f"item_{i}_done", range(5))
    print(list(results))

# 🔲 BORDERLINE: ThreadPoolExecutor vs ProcessPoolExecutor
# Threads: good for I/O-bound (network, file, DB)
# Processes: good for CPU-bound (image proc, ML, parsing)
# GIL limits true parallelism for CPU work in threads
```

```python
# --- Producer-Consumer with Queue (thread-safe) ---
# ✅ Custom: classic pattern with graceful shutdown
import threading
import queue
import time
import random

def producer(q, items, stop_event):
    for item in items:
        time.sleep(random.uniform(0.05, 0.15))
        q.put(item)
        print(f"  [Producer] queued: {item}")
    # signal consumers to stop
    stop_event.set()
    print("  [Producer] done, stop event set")

def consumer(name, q, stop_event):
    while True:
        try:
            item = q.get(timeout=0.3)   # wait max 300ms per attempt
            print(f"  [{name}] processed: {item}")
            q.task_done()
        except queue.Empty:
            if stop_event.is_set():
                print(f"  [{name}] shutting down")
                return

q = queue.Queue(maxsize=5)              # bounded buffer
stop_event = threading.Event()

items = [f"job_{i}" for i in range(8)]
prod = threading.Thread(target=producer, args=(q, items, stop_event))
cons1 = threading.Thread(target=consumer, args=("C1", q, stop_event))
cons2 = threading.Thread(target=consumer, args=("C2", q, stop_event))

for t in [prod, cons1, cons2]: t.start()
for t in [prod, cons1, cons2]: t.join()

print("All done. Queue empty:", q.empty())

# ⚠️ TRAP: Never use list/deque as a queue across threads -- not thread-safe
# ✅ queue.Queue is designed for multi-threaded producer-consumer
```

```python
# --- Threading with daemon threads + Timer ---
# ✅ Custom: background monitor pattern

import threading
import time

class HealthMonitor:
    def __init__(self, interval=0.5):
        self.interval = interval
        self.running = False
        self._thread = None
        self.checks = 0

    def _run(self):
        while self.running:
            self.checks += 1
            # In real use: check disk, memory, network etc.
            print(f"  [Monitor] health check #{self.checks}")
            time.sleep(self.interval)

    def start(self):
        self.running = True
        self._thread = threading.Thread(target=self._run, daemon=True)
        # daemon=True: thread dies when main program exits
        self._thread.start()
        print("[Monitor] started (daemon thread)")

    def stop(self):
        self.running = False
        self._thread.join(timeout=2)
        print(f"[Monitor] stopped after {self.checks} checks")

monitor = HealthMonitor(interval=0.3)
monitor.start()

# Main program does work
time.sleep(1.2)

monitor.stop()

# --- threading.Timer: run once after delay ---
def reminder(msg):
    print(f"[Timer] Reminder: {msg}")

t = threading.Timer(0.5, reminder, args=("5-minute break!",))
t.start()
t.join()   # wait for it

# ⚠️ TRAP: daemon threads are killed abruptly on exit -- don't do file I/O in them
```

```python
# --- SQLite: context manager for connection + transaction rollback ---
# custon: safe DB connection context manager
import sqlite3
from contextlib import contextmanager

DB_PATH = ':memory:'   # in-memory for demo

@contextmanager
def get_db(path=DB_PATH):
    conn = sqlite3.connect(path)
    conn.row_factory = sqlite3.Row
    try:
        yield conn
        conn.commit()       # auto-commit on clean exit
    except Exception:
        conn.rollback()     # auto-rollback on any error
        raise
    finally:
        conn.close()

# Setup
with get_db() as conn:
    conn.execute(
        'CREATE TABLE users '
        '(id INTEGER PRIMARY KEY AUTOINCREMENT, '
        ' name TEXT NOT NULL, email TEXT UNIQUE NOT NULL, age INTEGER)'
    )
    conn.executemany(
        'INSERT INTO users (name, email, age) VALUES (?,?,?)',
        [('Alice','alice@example.com',30),
         ('Bob','bob@example.com',25),
         ('Charlie','charlie@example.com',35)]
    )
    print('Setup complete')

# Query
with get_db() as conn:
    rows = conn.execute('SELECT * FROM users WHERE age > ?', (26,)).fetchall()
    for r in rows:
        print(dict(r))

# Rollback demo -- UNIQUE violation causes full rollback
try:
    with get_db() as conn:
        conn.execute('INSERT INTO users (name,email,age) VALUES (?,?,?)', ('Dave','dave@ex.com',28))
        conn.execute('INSERT INTO users (name,email,age) VALUES (?,?,?)', ('Alice','alice@example.com',99))  # UNIQUE violation
except sqlite3.IntegrityError as e:
    print(f'Rolled back: {e}')

# Dave should NOT appear (rolled back)
with get_db() as conn:
    c = conn.execute("SELECT COUNT(*) FROM users WHERE name='Dave'").fetchone()
    print('Dave in DB after rollback?', c[0])  # 0

# TRAP: without rollback, partial inserts stay committed -- always use context manager!
```

```python
# --- ORM-style repository pattern ---
import sqlite3
from dataclasses import dataclass
from typing import Optional

@dataclass
class Product:
    name:  str
    price: float
    stock: int
    id:    Optional[int] = None

class ProductRepo:
    def __init__(self, conn):
        self.conn = conn
        conn.row_factory = sqlite3.Row
        conn.execute(
            'CREATE TABLE IF NOT EXISTS products '
            '(id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, price REAL, stock INTEGER)'
        )
        conn.commit()

    def add(self, p):
        cur = self.conn.execute(
            'INSERT INTO products (name, price, stock) VALUES (?,?,?)',
            (p.name, p.price, p.stock)
        )
        self.conn.commit()
        p.id = cur.lastrowid
        return p

    def find(self, pid):
        row = self.conn.execute('SELECT * FROM products WHERE id=?', (pid,)).fetchone()
        return Product(**dict(row)) if row else None

    def update_stock(self, pid, delta):
        self.conn.execute('UPDATE products SET stock = stock + ? WHERE id=?', (delta, pid))
        self.conn.commit()

    def all(self):
        rows = self.conn.execute('SELECT * FROM products ORDER BY price').fetchall()
        return [Product(**dict(r)) for r in rows]

conn = sqlite3.connect(':memory:')
repo = ProductRepo(conn)

p1 = repo.add(Product('Apple',  1.20, 100))
p2 = repo.add(Product('Banana', 0.50, 200))
p3 = repo.add(Product('Cherry', 3.00,  50))

print('All products:')
for p in repo.all():
    print(f'  [{p.id}] {p.name:10} ${p.price:.2f}  stock={p.stock}')

repo.update_stock(p3.id, -10)
print('Cherry after sale:', repo.find(p3.id))
conn.close()
```

```python
# --- SQL injection prevention: always use ? placeholders ---
# TRAP: NEVER use f-strings or .format() in SQL queries
import sqlite3

conn = sqlite3.connect(':memory:')
conn.execute('CREATE TABLE users (id INTEGER PRIMARY KEY, name TEXT, role TEXT)')
conn.executemany('INSERT INTO users VALUES (?,?,?)',
    [(1,'Alice','admin'),(2,'Bob','user'),(3,'Charlie','user')])
conn.commit()

# SAFE -- parameterized
def safe_get_user(conn, name):
    return conn.execute('SELECT * FROM users WHERE name=?', (name,)).fetchone()

# UNSAFE -- injection vulnerable (demo only, never do this!)
def unsafe_get_user(conn, name):
    return conn.execute(f"SELECT * FROM users WHERE name='{name}'").fetchall()

print('Safe lookup:', safe_get_user(conn, 'Alice'))
print('Safe with injection attempt:', safe_get_user(conn, "' OR '1'='1"))

# Injection dumps entire table!
malicious = "' OR '1'='1"
print('UNSAFE result (all rows leaked):', unsafe_get_user(conn, malicious))

# executemany vs loop -- executemany is faster
import time
data = [(i,) for i in range(1000)]

c1 = sqlite3.connect(':memory:')
c1.execute('CREATE TABLE t (x INTEGER)')
t0 = time.perf_counter()
for row in data: c1.execute('INSERT INTO t VALUES (?)', row)
c1.commit()
t1 = time.perf_counter()

c2 = sqlite3.connect(':memory:')
c2.execute('CREATE TABLE t (x INTEGER)')
t2 = time.perf_counter()
c2.executemany('INSERT INTO t VALUES (?)', data)
c2.commit()
t3 = time.perf_counter()

print(f'Loop:        {(t1-t0)*1000:.1f} ms')
print(f'executemany: {(t3-t2)*1000:.1f} ms')
```

```python
# --- JOINs + Aggregate queries ---
import sqlite3

conn = sqlite3.connect(':memory:')
conn.row_factory = sqlite3.Row

conn.executescript(
    'CREATE TABLE categories (id INTEGER PRIMARY KEY, name TEXT);'
    'CREATE TABLE products (id INTEGER PRIMARY KEY, name TEXT, price REAL, cat_id INTEGER);'
    'CREATE TABLE sales (id INTEGER PRIMARY KEY, prod_id INTEGER, qty INTEGER, date TEXT);'
)

conn.executemany('INSERT INTO categories VALUES (?,?)',
    [(1,'Fruits'),(2,'Dairy'),(3,'Bakery')])
conn.executemany('INSERT INTO products VALUES (?,?,?,?)',
    [(1,'Apple',1.2,1),(2,'Banana',0.5,1),(3,'Milk',1.8,2),(4,'Bread',2.5,3)])
conn.executemany('INSERT INTO sales VALUES (?,?,?,?)',
    [(1,1,10,'2024-03-01'),(2,2,5,'2024-03-01'),
     (3,1,8,'2024-03-02'),(4,3,12,'2024-03-02'),
     (5,4,3,'2024-03-02'),(6,1,15,'2024-03-03')])
conn.commit()

print('=== Revenue by product ===')
rows = conn.execute(
    'SELECT p.name, SUM(s.qty) AS total_qty, ROUND(SUM(s.qty*p.price),2) AS revenue '
    'FROM sales s JOIN products p ON s.prod_id=p.id '
    'GROUP BY p.id ORDER BY revenue DESC'
).fetchall()
for r in rows:
    print(f"  {r['name']:10} qty={r['total_qty']:3}  revenue=${r['revenue']:.2f}")

print()
print('=== Revenue by category ===')
rows = conn.execute(
    'SELECT c.name, ROUND(SUM(s.qty*p.price),2) AS revenue '
    'FROM sales s JOIN products p ON s.prod_id=p.id '
    'JOIN categories c ON p.cat_id=c.id '
    'GROUP BY c.id ORDER BY revenue DESC'
).fetchall()
for r in rows:
    print(f"  {r['name']:10} ${r['revenue']:.2f}")

conn.close()
```
